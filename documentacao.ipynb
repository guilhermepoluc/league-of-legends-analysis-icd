# Introdução

League of Legends é um jogo de estratégia em que duas equipes escolhem cinco Campeões cada e se enfrentam para destruir a base uma da outra. O jogo possui mais de 140 Campeões para escolher, onde seu objetivo é destruir o "Nexus" inimigo. Este jogo possui um cenário competitivo muito forte e ativo, no qual existem campeonatos nacionais em vários países, além de um campeonato mundial que reune as melhores equipes do ano específico.

Este trabalho tem como objetivo realizar uma análise sobre os atributos e métricas que envolvem as partidas do jogo League of Legends. Para isso, foram utilizadas três bases de dados, capturadas da Kaggle, onde serão explicadas mais abaixo neste documento.

## Motivação
Como ambos os integrantes do grupo jogam League of Legends rotineiramente, o interesse em realizar este tipo de análise mais científica e técnica foi uma grande motivação para implementar o trabalho. Além disso, podemos atestar de forma experimental alguns dos mitos e verdades sobre o contexto do jogo, como realização de objetivos, participação em abates e taxas de vitória.

## Perguntas da Pesquisa
- Objetivos são importantes para vencer uma partida?
- Eliminar jogadores inimigos é importante para vencer uma partida?
- O acesso a visão de mapa é importante para vencer uma partida?
- O “first blood” é determinante para a vitória da equipe?

## Descrição da base

Pesquisamos algumas bases de dados para que tivessem bastante conteúdo para explorar e encontramos três que satisfazem nosso planejamento inicial sobre como seria e qual a análise que faríamos.
- 1º Base: https://www.kaggle.com/datasnaek/league-of-legends?select=games.csv

Esta é um conjunto que reune mais de 50000 partidas ranqueadas de League of Legends. Possui 61 atributos (colunas), nos quais as mais importantes são:
- Duração da partida
- Ganhador da partida (1 = equipe 1, 2 = equipe 2)
- Primeira realização de objetivos (Barão, dragão, torre, inibidor, "blood" e Herald)
- Campeões e magias selecionadas pelos times
- Número de objetivos feitos em cada time
- Campeões banidos em cada time na partida 

- 2º base: https://www.kaggle.com/chuckephron/leagueoflegends?select=LeagueofLegends.csv

Esta base possui as partidas competitivas, das ligas NALCS, EULCS, LCK, LMS, and CBLoL, assim como o campeonato mundial e o torneio da meia temporada, entre os anos de 2015 e 2017. Possui 57 colunas bem detalhadas, onde há informações sobre a quantidade de gold, banimento e escolha de campeões, kills e realização de objetivos para cada equipe, além de estar separada pelas funções de cada jogador na partida (Support, ADC, Middle, Jungle e Top).

- 3º base: https://www.kaggle.com/paololol/league-of-legends-ranked-matches?select=matches.csv

Largo conjunto de dados com mais de 180000 partidas ranqueadas presentes, na categoria Solo, espaçadas durante os anos. Seus dados estão organizados em basicamnete 3 tabelas com infromações sobre: Partidas, Jogador e estatística do time, além dos Banimentos.   

## Métodos e modelos utilizados

Testes de Hipotéses pensados:

- A quantidade de objetivos atingidos não influencia na taxa de vitória. (permutação com intervalo de confiança)
- A quantidade de kills feitos não influenciam na taxa de vitória.
- A quantidade de dano deferido a campeões pela equipe não influencia na taxa de vitória.
- A quantidade de dano deferido a objetivos não influencia na taxa de vitória.
- O farm não influencia na taxa de vitória.
- A quantidade de sentinelas utilizadas não influencia na taxa de vitória.
- O “first blood” não influencia na taxa de vitória da equipe.
