# Introdução

League of Legends é um jogo de estratégia em que duas equipes escolhem cinco Campeões cada e se enfrentam para destruir a base uma da outra. O jogo possui mais de 140 Campeões para escolher, onde seu objetivo é destruir o "Nexus" inimigo. Este jogo possui um cenário competitivo muito forte e ativo, no qual existem campeonatos nacionais em vários países, além de um campeonato mundial que reune as melhores equipes do ano específico.

Este trabalho tem como objetivo realizar uma análise sobre os principais atributos e métricas que envolvem as partidas do jogo League of Legends. Para isso, foram utilizadas três bases de dados, capturadas da Kaggle, onde serão explicadas mais abaixo neste documento.

## Motivação
Como ambos os integrantes do grupo jogam League of Legends rotineiramente, o interesse em realizar este tipo de análise mais científica e técnica foi uma grande motivação para implementar o trabalho. Além disso, podemos atestar de forma experimental alguns dos mitos e verdades sobre o contexto do jogo, como realização de objetivos, participação em abates e taxas de vitória.

## Descrição da base

Pesquisamos algumas bases de dados para que tivessem bastante conteúdo para explorar e encontramos três que satisfazem nosso planejamento inicial sobre como seria e qual a análise que faríamos.
- 1º Base: https://www.kaggle.com/datasnaek/league-of-legends?select=games.csv

Esta é um conjunto que reune mais de 50000 partidas ranqueadas de League of Legends. Possui 61 atributos (colunas), nos quais as mais importantes são:
- Duração da partida
- Ganhador da partida (1 = equipe 1, 2 = equipe 2)
- Primeira realização de objetivos (Barão, dragão, torre, inibidor, "blood" e Herald)
- Campeões e magias selecionadas pelos times
- Número de objetivos feitos em cada time
- Campeões banidos em cada time na partida 


- 2º base: https://www.kaggle.com/chuckephron/leagueoflegends?select=LeagueofLegends.csv

Esta base possui as partidas competitivas, das ligas NALCS, EULCS, LCK, LMS, and CBLoL, assim como o campeonato mundial e o torneio da meia temporada, entre os anos de 2015 e 2017. Possui 57 colunas bem detalhadas, onde há informações sobre:
- Quantidade de gold, 
- Banimento e escolha de campeões,
- Kills,
- Realização de objetivos para cada equipe, além de estar separada pelas funções de cada jogador na partida (Support, ADC, Middle, Jungle e Top).


- 3º base: https://www.kaggle.com/paololol/league-of-legends-ranked-matches?select=matches.csv

Largo conjunto de dados com mais de 180000 partidas ranqueadas presentes, na categoria Solo, espaçadas durante os anos. Seus dados estão organizados em basicamente 3 tabelas com informações sobre: Partidas, Jogador e estatística do time, além dos Banimentos. Nesta base focamos na tabela sobre os dados dos jogadores, onde contêm 56 colunas extremamente detalhadas, como por exemplo informações sobre:
- Itens adquiridos
- Gold ganho e gasto
- Dano sofrido e gerado, tanto em torres quanto de inimigos
- Kills, assistências e mortes
- Wards compradas, quebradas e colocadas (variável relacionada a visualização de mapa)
- Total de minions eliminados


## Métodos e modelos utilizados

Testes de Hipotéses pensados:

- A quantidade de objetivos atingidos não influencia na taxa de vitória. (permutação com intervalo de confiança)
- A quantidade de kills feitos não influenciam na taxa de vitória.
- A quantidade de dano deferido a campeões pela equipe não influencia na taxa de vitória.
- A quantidade de dano deferido a objetivos não influencia na taxa de vitória.
- O farm não influencia na taxa de vitória.
- A quantidade de sentinelas utilizadas não influencia na taxa de vitória.
- O “first blood” não influencia na taxa de vitória da equipe.

Decidimos gerar os intervalos de confiança para os testes de hipótese utilizando a técnica de permutação, na qual aleatorizados um dos atributos para entender como seria a distribuição dos dados em um cenário onde as associações são aleatórias. Por exemplo, inicialmente comparamos a diferença média de conquista de objetivo entre a equipe vencedora e a perdedora, essa é a taxa observada na base de dados original. Depois aleatorizamos quais foram as equipes que venceram e as que perderam nos registros e calculamos a mesma taxa. Repetimos essa última etapa várias vezes e plotamos um histograma das taxas observadas nesse cenário. Essa distribuição que obtemos servirá de base para compararmos a taxa observada inicialmente com essa nova distribuição que representa um cenário em que os objetivos não interferem na taxa de vitória. Caso nossa taxa encontrada inicialmente esteja dentro do intervalo de confiança, aceitamos a hipótese nula, caso esteja fora, rejeitamos, a partir de um determinado nível de significância.

## Previsão

Esta parte foi dividida em regressão e classificação, no qual pensamos em duas propostas para cada, que serão explicadas com mais detalhes abaixo. Como foi comentado anteriormente, os arquivos **regressao.ipynb** e **classificacao.ipynb** possuem as implementações, de forma explicada e comentada, das respectivas partes.

Para ambas as partes, os fluxos dos notebooks são da seguinte forma:

### Regressão

Propôs-se as seguintes regressões:
- Estimar o gold do jogador na partida com o número de objetivos feitos (torres e inibidores destruídos, dragões e barões eliminados), total de minions eliminados, kills feitas, além das wards eliminadas.
- Estimar a quantidade de kills, assistências e mortes de um jogador, de acordo com o gold gasto.

Na 1ª regressão, com o objetivo de realizar uma análise do comportamento das variáveis no resultado final, dividimos em 3 experimentos, onde:
- No 1º queremos estimar o gold do jogador somente utilizando o número de kills
- No 2º estimar o gold somente utilizando o número de minions eliminados
- No 3º estimar o gold utilizando todas as variáveis propostas no regressão inicial

Os resultados desta regressão podem ser encontrados em seu respectivo arquivo, onde podemos já afirmar que o erro quadrático médio (métrica utilizada para avaliar a operação) é nitidamente menor para o 3º experimento. Além disso, no arquivo, plotamos as retas aproximadas dos experimentos 1 e 2 para avaliar se a aproximação estava acontecendo de forma correta. Este processo está ilustrado no arquivo desta parte. 

Já para a 2ª regressão, podemos analisar que as retas traçadas para as 3 aproximações foram feitas de forma correta, apesar do alto erro quadrático médio gerado. A partir dos gráficos podemos ver que a reta consegue aproximar pouco o comportamento das variáveis escolhidas. O resultado não é tão significativo pois as amostras se encontram bem esparsas e uma reta em si não conseguiria aproximá-las da forma ideal.  


### Classificação

Para esta seção foram pensadas em duas classificações para aplicar nos dados:
- Classificar o vencedor baseado em objetivos feitos.
- Classificar o vencedor baseado nos primeiros objetivos feitos.

Estas propostas foram pensadas para de fato responder se os objetivos são realmente importante para determinar o vencedor, a ponto de classificá-lo no final da partida e logo quando os primeiros objetivos são executados.

Para ambos os casos, as acurácias foram acima de 90%, indicando que o algoritmo conseguiu identificar o vencedor de acordo com as variáveis escolhidas. Esta análise pode ser conferida no respectivo documento desta parte (classificacao.ipynb).
